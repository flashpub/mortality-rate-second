# Mortality Trends in Switzerland and Japan: A Comparative Analysis Over the Past 60 Years (1960-2019)

## Introduction

Mortality, as a key indicator of public health, is an area of research that has a profound impact on people's lives and society as a whole. Monitoring trends in mortality is an important part of public health analysis and allows assessing the effectiveness of the health system, risk factors and changes in the health of the population in the future. In this study, we focus on a comparative analysis of mortality in two different countries, Switzerland and Japan, over the past 60 years. By analyzing data on gender and age groups, we aim to identify common patterns and differences that can shed light on important aspects of health and the health system in these countries. This research is of critical importance for the development of healthcare and the development of measures aimed at improving the quality of life of the population in the future.

This leads to our research question: **How do the mortality trends for Switzerland and Japan differ over the past 60 years, depending on age, gender and cause of death?**

The selection of Switzerland and Japan as the focal countries for this comparative analysis stems from a deliberate consideration of their contrasting healthcare system types and distinct geographic locations. Switzerland has a robust private healthcare system contrasts while Japan has comprehensive universal healthcare system. Covering the span of 1960 to 2019  is driven by compelling historical and medical factors. The 1960s marked a turning point in the realm of medicine. It was a period when the world had largely from the aftermath of the Second World War, and the emphasis shifted from wartime-related fatalities to significant advancements in medical science. This era witnessed rapid developments in medical technologies and treatments, addressing the persistently high mortality rates attributed to diseases. Additionally, our approach considers the crucial dimension of age groups and gender disparities, thereby avoiding overgeneralization by recognizing the diversity within the populations under study. The selected time frame of 1960 to 2019 also ensures that the research is not skewed by the exceptional circumstances of the COVID-19 pandemic, which could disproportionately affect mortality statistics during the subsequent years. This comprehensive framework offers a multifaceted perspective on mortality trends.

## Data and Methodology

**Data Source:** The primary data source for this research is the World Health Organization's Global Health Observatory (GHO), which provides comprehensive mortality data. Specifically, the data was retrieved from the GHO's Mortality Database (MDB) for all causes of death. The dataset can be accessed via the following link: Mortality Database - All Causes.

**Data Cleaning Process:** To ensure the integrity and reliability of the dataset, a rigorous data cleaning process was undertaken. This process involved the identification and removal of missing values (NAs) and corrupted data points. By eliminating such data artifacts, the dataset was prepared for further analysis.

**Variables Included in the Datasets:** The dataset under examination consists of multiple variables, providing a comprehensive view of mortality statistics. These variables include:

Region Code
Region Name
Country Code
Country Name
Year
Sex
Age Group Code
Age Group
Number
Percentage of cause-specific deaths out of total deaths
Age-Standardized Death Rate per 100,000 standard population
Death Rate per 100,000 population

**Datasets for Different Causes:** The research encompassed multiple datasets, each focused on specific causes of death, including "All causes," "Communicable, maternal, perinatal and nutritional conditions," "Noncommunicable diseases," "Injuries," and "Ill-defined diseases." These datasets provided a nuanced examination of mortality trends across various causes.

**Statistical Analysis:** The dataset underwent a comprehensive statistical analysis. Descriptive statistics, including measures such as mean, median, standard deviation, and quartiles, were calculated using the describe function to gain insights into the dataset's central tendencies and variability.

**Distribution Assessment:** The variables in the dataset were scrutinized for their distribution characteristics. The Shapiro-Wilk test was applied to assess variables for normality and log-normality, while nonparametric tests were employed for other distributions.

**Testing Gender Differences:** To explore potential disparities between genders, statistical tests were conducted on the numeric variables. Depending on the distribution of the data, either t-tests or the Wilcoxon-Mann-Whitney criterion were employed to test for significant differences between genders.

**Data Visualization:** Data visualization played a pivotal role in this research. Various Python visualization tools and libraries were used to create graphs and charts, which enhanced the presentation of mortality trends by gender and age groups, and their relationships with specific causes of death.
